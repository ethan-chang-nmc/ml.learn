```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('deeplearning.mplstyle')

# x_train as input variable (size in 1000 square feet)
# y_train as target (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}\ny_train = {y_train}")
```

```python
# m is the number of training examples (1D array (vector) so it returns 1 value)
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}") # can also use m = len(x_train)
```

```python
# to see i^{th} training example
for i in range(m):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"(x^({i}), y^({i})) = ({x_i}, {y_i}))")
```

```python
# plot the data points
plt.scatter(x_train, y_train, marker="x", c="r")
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollar)")
plt.xlabel("Size (1000 sqft")
plt.show()
```

#### Model Function

Model function for linear regression is given by
$$f_{w,b}(x^{(i)}) = wx^{(i)} + b$$
We want to optimize parameter values of $w$ and $b$. Can generalize to:
$$f_{w_{k, k = 1, 2, \dots},b}(x_{1}^{(i_1)}, x_{2}^{(i_2)}, \dots, x_k^{(i_k)}) = w_{1}x_{1}^{(i_1)} + w_{2}x_{2}^{(i_2)} + \dots + w_{k}x_{k}^{(i_k)} + b$$

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
        x (ndarray (m,)): Data, m examles
        w, b (scalar)   : model parameters
    Returns
        f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, tmp_f_wb, c = "b", label = "Our Prediction")
plt.scatter(x_train, y_train, marker = "x", c = "r", label = "Actual Values")

plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.legend()
plt.show()
```
